```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.value_counts('Outcome')
```

```python
sns.lineplot(data=df, x='Age', y='BloodPressure')
```

```python
sns.histplot(data=df, x='Age', bins=7, kde=True)
```

```python
sns.kdeplot(data=df, x='Age')
```

```python
sns.boxplot(data=df)
```

```python
df['Resultado'] = ''
```

```python
df.loc[df['Outcome'] == 1, 'Resultado'] = 'Diabético'
df.loc[df['Outcome'] == 0, 'Resultado'] = 'Não Diabético'
sns.countplot(x=df['Resultado'], palette='Blues')
```

```python
sns.scatterplot(data=df, x='Insulin', y='Glucose')
```
